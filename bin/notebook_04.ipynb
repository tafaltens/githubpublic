## Another way to display the first kitten using os.path.join

This shows another way to represent the path to the first kitten.  
This path will be stored as the variable ```kitten1_file```.

````python
# import the libraries for operating system functions and displaying images.
import os
from IPython.display import Image
````

````python
# create the path to the kitten_1.png file using os.path.join
# save this path as the variable kitten01_file and print the path.
kitten1_file = os.path.join('..','examples','kitten_1.png')
kitten1_file
````

To display the kitten, use the Image command with the variable for the path:
```Image(kitten1_file)```



````python
Image(kitten1_file) 
````
