````python
# Import the libraries to be used in this notebook
import glob
from IPython.display import Image, display
````

## Search for all images in the examples folder 

````python
# Store the filenames in a list called all_image_locations.
all_image_locations=[]
for x in glob.glob('examples/*.png', recursive=True):
    print(x)
    all_image_locations.append(x)
````

## Search for all images in the examples folder and all of its subfolders.

````python
# Store the filenames in a list called all_image_locations.
all_image_locations=[]
for x in glob.glob('examples/**/*.png', recursive=True):
    print(x)
    all_image_locations.append(x)
````

## Search for all kittens in all folders in this app.

````python
# Create a list to store the locations of all the kittens.
all_kitten_locations = []
for file_path in glob.glob('**/kitten*.png', recursive=True): # recursively search through all the .png files for filenames containing kitten.
    all_kitten_locations.append(file_path)  # append new kitten locations to the list
# sort the results - but this includes the path name
all_kitten_locations.sort()
display(all_kitten_locations)
````

````python
# Display the kitten images
for item in all_kitten_locations:
    display(Image(filename=item)) # display the image of the kitten
    display(item)                 # also display the pathname for the kitten 
````

### Exercise: Search for  and print all .png images in this app, along with their locations

````python
# Edit the code below to search for and display ALL of the images in this app, not just the kittens.
all_image_locations=[]

for image_file in glob.glob('**/kitten*.png', recursive=True):
    all_image_locations.append(image_file)
    
# Display all of the images
for image_loc in all_image_locations:
    display(Image(filename=image_loc)) # display the image - not only kittens!
    display(image_loc)                 # also display the pathname for the image. 
````
