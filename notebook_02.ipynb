Each nanoHUB App template is a set of 7 folders plus an invoke script. Some details about these folders can be found on the [GettingStarted page in nanoFORGE](https://nanohub.org/tools/githubpublic/wiki/GettingStarted) for this app.

## The 7 nanoHUB App Template Folders 

 * <b>bin</b> - binaries go here.  The invoked file can also be in this folder.  Some people put all of their notebooks and related files here.
 * <b>data</b> - optional folder.  Data files such as .csv or .txt files that are utilized in the tool can be organized here.
 * <b>doc</b> - optional folder.  Documentation can be put here.
 * <b>examples</b> - optional folder.  Example notebooks can be put here.
 * <b>middleware</b> - the executable invoke script goes here.  The app template has an invoke script template with the correct access permissions. 
 * <b>rappture</b> - this folder can remain unused for Jupyter notebook tools.  
 * <b>src</b> - code that must be compiled should reside here. The src should include a Makefile for building the tool. Products of the build such as binary executables should be placed the top level directory named bin.
 
Additional folders can also be created and used. 
 
Important: The file that is invoked from the invoke script is first searched for at the top level of the app or in the bin folder. It may be best to put this file at the top level of the app because only files that are in the same folder or in subfolders of the invoked file can be accessed via the Jupyter Dashboard. If you place the invoked file within a subfolder, the files in other folders may not be viewable from the Jupyter Dashboard. These files, however, can still be accessed by using their path from within a notebook.

### Source Code Browser in nanoFORGE

You can view the actual template folders and other files for this app in the [Source Code browser in nanoFORGE](https://nanohub.org/tools/githubpublic/browser).



Once you have set up your GitHub tool repo, you can register your new app in nanoHUB. 

Keep in mind that even though you are using GitHub as the app repo, your app will still need to be tested in your nanoHUB file space, then pushed to nanoFORGE (uploaded) and nanoHUB Apps (installed) where you will test it again before requesting that the app be published.

Be mindful that through this process, different versions of your code may be in the various code repositories and related places.  These include:

   * your local repo in your nanoHUB filespace
   * the external repo in your GitHub account
   * the nanoFORGE repo, where you can Browse the Source code, download the code, and view the nanoFORGE revision history for the app.
   * nanoHUB apps, where the published tool and previously published versions of the code live.
   * the tarball that can be downloaded from the tool's page in nanoHUB. For this tool, the tool page will be: https://nanohub.org/tools/githubpublic
